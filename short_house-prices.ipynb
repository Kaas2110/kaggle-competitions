```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Regressors
from xgboost import XGBRegressor

from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
import lightgbm as lgb
```

```python

Data_train=pd.read_csv('train.csv')
Data_test=pd.read_csv('test12.csv')

train=Data_train.copy()
test=Data_test.copy()

cols=["Id","MiscFeature","Fence"
      ,"HouseStyle","RoofStyle","RoofMatl","Exterior1st",
      "Exterior2nd","MasVnrType","Foundation"]#

train=train.drop(columns=cols)
test=test.drop(columns=cols)
```

```python

train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))

test["LotFrontage"] = test.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))

train=train.fillna(0)
test=test.fillna(0)

qual={'Wall':0,'OthW':1,'Grav':2,'GasW':3,'GasA':4,'Floor':5}
train.Heating.replace(qual,inplace=True)
test.Heating.replace(qual,inplace=True)

qualities = {'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5}

train.FireplaceQu.replace(qualities,inplace=True)
test.FireplaceQu.replace(qualities,inplace=True)
train.GarageQual.replace(qualities,inplace=True)
test.GarageQual.replace(qualities,inplace=True)
train.GarageCond.replace(qualities,inplace=True)
test.GarageCond.replace(qualities,inplace=True)
train.BsmtQual.replace(qualities,inplace=True)
test.BsmtQual.replace(qualities,inplace=True)
train.BsmtCond.replace(qualities,inplace=True)
test.BsmtCond.replace(qualities,inplace=True)
train.ExterQual.replace(qualities,inplace=True)
test.ExterQual.replace(qualities,inplace=True)
train.ExterCond.replace(qualities,inplace=True)
test.ExterCond.replace(qualities,inplace=True)
train.HeatingQC.replace(qualities,inplace=True)
test.HeatingQC.replace(qualities,inplace=True)
train.KitchenQual.replace(qualities,inplace=True)
test.KitchenQual.replace(qualities,inplace=True)
train.PoolQC.replace(qualities,inplace=True)
test.PoolQC.replace(qualities,inplace=True)

qualities = {'NA':0,'No':1,'Mn':2,'Av':3,'Gd':4}
train.BsmtExposure.replace(qualities,inplace=True)
test.BsmtExposure.replace(qualities,inplace=True)

qualities = {'NA':0,'Unf':1,'LwQ':2,'Rec':3,'BLQ':4,'ALQ':5,'GLQ':6}
train.BsmtFinType1.replace(qualities,inplace=True)
test.BsmtFinType1.replace(qualities,inplace=True)
train.BsmtFinType2.replace(qualities,inplace=True)
test.BsmtFinType2.replace(qualities,inplace=True)

G_parms = {'Detchd':1,'CarPort':2,'BuiltIn':3,'Basment':4,'Attchd':5,'2Types':6}
train.GarageType.replace(G_parms,inplace=True)
test.GarageType.replace(G_parms,inplace=True)

G_parms = {'Unf':1,'RFn':2,'Fin':3}
train.GarageFinish.replace(G_parms,inplace=True)
test.GarageFinish.replace(G_parms,inplace=True)

qualities = {'NA':0,'RM':1,'RP':2,'RL':3,'RH':4,'I':5,'FV':6,'C (all)':7,'A':8}
train.MSZoning.replace(qualities,inplace=True)
test.MSZoning.replace(qualities,inplace=True)

G_parms = {'Pave':1,'Grvl':2}
train.Street.replace(G_parms,inplace=True)
test.Street.replace(G_parms,inplace=True)
train.Alley.replace(G_parms,inplace=True)
test.Alley.replace(G_parms,inplace=True)

G_parms = {'IR3':0,'IR2':1,'IR1':2,'Reg':3}
train.LotShape.replace(G_parms,inplace=True)
test.LotShape.replace(G_parms,inplace=True)

G_parms = {'Low':0,'HLS':1,'Bnk':2,'Lvl':3}
train.LandContour.replace(G_parms,inplace=True)
test.LandContour.replace(G_parms,inplace=True)

G_parms = {'ELO':0,'NoSeWa':1,'NoSewr':2,'AllPub':3}
train.Utilities.replace(G_parms,inplace=True)
test.Utilities.replace(G_parms,inplace=True)

G_parms = {'FR3':0,'FR2':1,'CulDSac':2,'Corner':3,'Inside':4}
train.LotConfig.replace(G_parms,inplace=True)
test.LotConfig.replace(G_parms,inplace=True)

G_parms = {'Sev':0,'Mod':1,'Gtl':2}
train.LandSlope.replace(G_parms,inplace=True)
test.LandSlope.replace(G_parms,inplace=True)

G_parms = {'Veenker':0,'Timber':1,'StoneBr':2,'Somerst':3,'SawyerW':4,'Sawyer':5,
           'SWISU':6,'OldTown':7,'NWAmes':8,'NridgHt':9,'NPkVill':10,'NoRidge':11,
           'NAmes':12,'Mitchel':13,'MeadowV':14,'IDOTRR':15,'Gilbert':16,'Edwards':17,
           'Crawfor':18,'CollgCr':19,'ClearCr':20,'BrkSide':21,'BrDale':22,'Blueste':23,
           'Blmngtn':24}
train.Neighborhood.replace(G_parms,inplace=True)
test.Neighborhood.replace(G_parms,inplace=True)

G_parms = {'RRAe':0,'RRNe':1,'PosA':2,'PosN':3,'RRAn':4,
           'RRNn':5,'Norm':6,'Feedr':7,'Artery':8}
train.Condition1.replace(G_parms,inplace=True)
test.Condition1.replace(G_parms,inplace=True)
train.Condition2.replace(G_parms,inplace=True)
test.Condition2.replace(G_parms,inplace=True)

G_parms = {'Twnhs':0,'TwnhsE':1,'Duplex':2,'2fmCon':3,'1Fam':4}
train.BldgType.replace(G_parms,inplace=True)
test.BldgType.replace(G_parms,inplace=True)

G_parms = {'N':0,'Y':1}
train.CentralAir.replace(G_parms,inplace=True)
test.CentralAir.replace(G_parms,inplace=True)

G_parms = {'Sal':0,'Sev':1,'Maj2':2,'Maj1':3,'Mod':4,
           'Min2':5,'Min1':6,'Typ':7}
train.Functional.replace(G_parms,inplace=True)
test.Functional.replace(G_parms,inplace=True)

G_parms = {'N':0,'P':1,'Y':2}
train.PavedDrive.replace(G_parms,inplace=True)
test.PavedDrive.replace(G_parms,inplace=True)

G_parms = {'Oth':0,'ConLD':1,'ConLI':2,'ConLw':3,'Con':4,
           'COD':5,'New':6,'VWD':7,'CWD':8,'WD':9}
train.SaleType.replace(G_parms,inplace=True)
test.SaleType.replace(G_parms,inplace=True)

G_parms = {'Partial':0,'Family':1,'Alloca':2,'AdjLand':3
           ,'Abnorml':4,'Normal':5}
train.SaleCondition.replace(G_parms,inplace=True)
test.SaleCondition.replace(G_parms,inplace=True)

G_parms = {'Mix':0,'FuseP':1,'FuseF':2,'FuseA':3
           ,'SBrkr':4}
train.Electrical.replace(G_parms,inplace=True)
test.Electrical.replace(G_parms,inplace=True)

train["Bathrooms"]=train["FullBath"]+(train["HalfBath"]*2)
train["Bsm_Bathrooms"]=train["BsmtFullBath"]+(train["BsmtHalfBath"]*2)

test["Bathrooms"]=test["FullBath"]+(test["HalfBath"]*2)
test["Bsm_Bathrooms"]=test["BsmtFullBath"]+(test["BsmtHalfBath"]*2)

train["LotArea"]=np.log(train["LotArea"])+1
test["LotArea"]=np.log(test["LotArea"])+1

train['TotalBath']=train['Bathrooms']+train['Bsm_Bathrooms']
test['TotalBath']=test['Bathrooms']+test['Bsm_Bathrooms']

train["OC_N_BAG"]=train['OverallQual']+train['Neighborhood']+train['Bathrooms']+train['BedroomAbvGr']
test["OC_N_BAG"]=test['OverallQual']+test['Neighborhood']+test['Bathrooms']+test['BedroomAbvGr']

train['SF_outside']=train['WoodDeckSF']+train['OpenPorchSF']
test['SF_outside']=test['WoodDeckSF']+test['OpenPorchSF']

train['TotalSF_Inside']=train['1stFlrSF']+train['2ndFlrSF']+train['TotalBsmtSF']
test['TotalSF_Inside']=test['1stFlrSF']+test['2ndFlrSF']+test['TotalBsmtSF']

train['TotBSmtFin']=train["BsmtFinType2"]+train["BsmtFinType1"]
test['TotBSmtFin']=test["BsmtFinType2"]+test["BsmtFinType1"]

train['NewQual']=(train['FireplaceQu']+train['PoolQC']+train['GarageQual']+train['ExterQual']+train['BsmtQual']+train['KitchenQual']+train['HeatingQC'])/7
test['NewQual']=(test['FireplaceQu']+test['PoolQC']+test['GarageQual']+test['ExterQual']+test['BsmtQual']+test['KitchenQual']+test['HeatingQC'])/7

train['NewCond']=(train['GarageCond']+train['BsmtCond']+train['ExterCond'])/3
test['NewCond']=(test['GarageCond']+test['BsmtCond']+test['ExterCond'])/3

train.loc[ train['YearBuilt'] ==train["YearRemodAdd"], 'Has_Remod'] = 0
train.loc[train['YearBuilt'] < train["YearRemodAdd"] ,'Has_Remod'] = 1

test.loc[ train['YearBuilt'] ==train["YearRemodAdd"], 'Has_Remod'] = 0
test.loc[train['YearBuilt'] < train["YearRemodAdd"] ,'Has_Remod'] = 1

train["MSSubClass"]=np.log1p(train["MSSubClass"])
test["MSSubClass"]=np.log1p(test["MSSubClass"])

train["TotalSF_Inside"]=np.log1p(train["TotalSF_Inside"])
test["TotalSF_Inside"]=np.log1p(test["TotalSF_Inside"])

train["SF_outside"]=np.log1p(train["SF_outside"])
test["SF_outside"]=np.log1p(test["SF_outside"])

train["NewQual"]=np.log1p(train["NewQual"])
test["NewQual"]=np.log1p(test["NewQual"])

train["LotFrontage"]=np.log1p(train["LotFrontage"])
test["LotFrontage"]=np.log1p(test["LotFrontage"])

train["TotBSmtFin"]=np.log1p(train["TotBSmtFin"])
test["TotBSmtFin"]=np.log1p(test["TotBSmtFin"])

cols=["Utilities",'GarageCond']



train=train.drop(columns=cols)
test=test.drop(columns=cols)
```

```python

from scipy import stats
print(train.shape)
z_scores=stats.zscore(train)

abs_z_scores=np.abs(z_scores)

filtered=(abs_z_scores<20).all(axis=1)
train=train[filtered]
print(train.shape)


#print(train.info())
X_train=train.drop(columns=["SalePrice"])
Y_train=train["SalePrice"]
X_test=test

cor_matrix=X_train.corr().abs()
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(to_drop)

X_train = X_train.drop(columns=to_drop, axis=1)
X_test = X_test.drop(columns=to_drop, axis=1)
```

```python

Xtr, Xte, Ytr, Yte = train_test_split(X_train, Y_train,test_size=0.01,random_state=52)


print("XGB test")
mdl = XGBRegressor(
   colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1
    ).fit(Xtr, Ytr)
print(mdl.score(Xte, Yte) )
xgb_predictions = mdl.predict(X_test)
#print(xgb_predictions)

output = pd.DataFrame({'Id': Data_test.Id, 'SalePrice': xgb_predictions})
output.to_csv('./House_XGB.csv', index=False)

print("GradientBoosting")
clf = GradientBoostingRegressor(
   n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
clf.fit(Xtr, Ytr)
print(clf.score(Xte, Yte) )
gb_predictions = clf.predict(X_test)
#print(xgb_predictions)
print("")

output = pd.DataFrame({'Id': Data_test.Id, 'SalePrice': gb_predictions})
output.to_csv('./House_GradienBoosting.csv', index=False)

print("")


from sklearn.ensemble import RandomForestRegressor
print("LGBM")
clf  = lgb.LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=720,
                              max_bin = 55, bagging_fraction = 0.8,
                              bagging_freq = 5, feature_fraction = 0.2319,
                              feature_fraction_seed=9, bagging_seed=9,
                              min_data_in_leaf =6, min_sum_hessian_in_leaf = 11)
clf.fit(Xtr, Ytr)
print(clf.score(Xte, Yte) )
rf_predictions = clf.predict(X_test)
#print(xgb_predictions)
print("")

output = pd.DataFrame({'Id': Data_test.Id, 'SalePrice': rf_predictions})
output.to_csv('./House_RandomForest.csv', index=False)


print("Average_Predictions")

#Arge=xgb_predictions*0.30+gb_predictions*0.70

Arge=xgb_predictions*0.30+gb_predictions*0.50+rf_predictions*0.2

output = pd.DataFrame({'Id': Data_test.Id, 'SalePrice': Arge})
output.to_csv('./Average.csv', index=False)
```
